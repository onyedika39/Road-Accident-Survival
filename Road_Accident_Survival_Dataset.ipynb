```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
accident = pd.read_csv('accident.csv')
print(accident)
```

```python
# information about the dataset
accident.info()
```

```python
accident.head()
```

```python
accident.dropna(inplace=True)
accident
```

```python
accident.dtypes
```

```python
# Summary Statistics
accident.describe()
```

```python
# Histogram For Age
plt.figure(figsize=(10,6))
accident['Age'].plot(kind = 'hist')
plt.title("Age Distribution.")
plt.show() 
```

```python
#Average age of the accident
print(accident["Age"].mean())
```

```python
#count the number of accidents by gender
print(accident["Gender"].value_counts())
```

```python
plt.figure(figsize =(6,4))
accident["Gender"].value_counts().plot(kind="bar", color=["blue", "pink"])
plt.title("Number of Accidents by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
                                       
```

```python
# find the number of people who survived and who did not
survival_counts = accident["Survived"].value_counts()
print(survival_counts)
```

```python
plt.figure(figsize = (6,6))
plt.pie(survival_counts,labels = ["Survived","Did Not Survive"],autopct='%1.1f%%',startangle=90)
plt.title("Survival Distribution")
plt.show()
```

```python
# find the maximum and minimum speed
max_speed = accident['Speed_of_Impact'].max()
min_speed = accident['Speed_of_Impact'].min()

print("Maximum Speed:", max_speed)
print("Minimum Speed:", min_speed)
```

```python
# Count the number of accidents where a helment was used
helmet_counts = accident["Helmet_Used"].value_counts()

# plotting the donut chart
plt.figure(figsize=(6, 6))
plt.pie(helmet_counts, labels=helmet_counts.index, autopct='%1.1F%%', startangle=90, pctdistance=0.85, colors=['#FFC107', '#2196F3'])
     
# Creating a white circle at the center to make it a donut chart
centre_center = plt.Circle((0,0), 0.70, color='white', fc='white')
plt.gca().add_artist(centre_center)
plt.title("Helmet Usage Distribution")
plt.show()
```

```python
# Count the number of accidents where a seatbelt was used
print(accident["Seatbelt_Used"].value_counts())
```

```python
# create a Bar chart for the count of Accident where a seatbelt was used
plt.figure(figsize=(6,4))
accident["Seatbelt_Used"].value_counts().plot(kind="bar",color=["green","red"])
plt.title("Number of Accidents by SeatBelt Usage")
plt.xlabel("Seatbelt Used")
plt.ylabel("Count")
plt.show()
```

```python
# percentage of male and female accident victims who survived
survival_rates = accident.groupby("Gender")["Survived"].mean() * 100

```

```python
print(survival_rates)
```

```python
# Bar chart showing the number of accidents by gender
accident["Gender"].value_counts().plot(kind="bar", color=["blue", "pink"])
plt.title("Number of Accidents by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
```

```python
# Pie chart showing the proportion of survivors and non-survivors
accident["Survived"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["red", "green"])
plt.title("Survival Rate")
plt.ylabel("")  # Remove default y-label
plt.show()
```

```python
# Find the survival rate of people who used a helmet
helmet_survival = accident.groupby("Helmet_Used")["Survived"].value_counts(normalize=True) * 100
print(helmet_survival)
```

```python
# Find the survival rate of people who used a seatbelt
seatbelt_survival = accident.groupby("Seatbelt_Used")["Survived"].value_counts(normalize=True) * 100
print(seatbelt_survival)
```
